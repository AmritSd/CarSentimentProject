```python
from transformers import pipeline

# Load the pre-trained sentiment analysis model
sentiment_analysis = pipeline("sentiment-analysis")

# Input comment for sentiment analysis
comment = "I love this product! It's amazing."

# Analyze sentiment of the comment
result = sentiment_analysis(comment)

# Extract sentiment label and score from the result
sentiment_label = result[0]['label']
sentiment_score = result[0]['score']

# Print the sentiment label and score
print("Sentiment Label:", sentiment_label)
print("Sentiment Score:", sentiment_score)
```

```python
import pandas as pd
import warnings

warnings.filterwarnings('ignore')

j = 0
def get_sentiment(x):
    global j
    j += 1
    if j % 1000 == 0:
        print(j)
    try:
        sentiment = sentiment_analysis(x)[0]
        return sentiment
    except Exception as e:
        return {'label': 'missing', 'score': 0.0}
    
for i in range(0, 3300, 100):
    start = i
    end = i + 100
    file_name = "./02_df_reviews_{}_{}_replaced.csv".format(start, end)
    try:
        df = pd.read_csv(file_name)
    except:
        continue

    
    # call get_sentiment in a loop
    df['sentiment'] = {'label': 'missing', 'score': 0.0}
    df['replaced_message'] = df['replaced_message'].astype(str)
    
    for k in range(0, len(df)):
        try:
            sentiment = get_sentiment(df['replaced_message'].iloc[k])
            df['sentiment'].iloc[k] = sentiment
        except:
            continue

    # df['sentiment_label'] = df['sentiment'].apply(lambda x: x['label'])
    # df['sentiment_score'] = df['sentiment'].apply(lambda x: x['score'])

    df.head()
    df.to_csv(file_name.replace('.csv', '_sentiment.csv'), index=False)
```
